```python
#another way to remove columns is by using drop:
#df.drop(['status'] axis=1) 
#the axis equals to 1 so that it knows it is dropping a column and not a row

df = df[['coaster_name', 
    #'Length', 'Speed', 
    'Location', 'Status', 
    #'Opening date',
    #   'Type',
    'Manufacturer',
    #'Height restriction', 'Model', 'Height',
    #   'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
   #    'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
   #    'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
    #   'Track layout', 'Fastrack available', 'Soft opening date.1',
    #   'Closing date', 
    'Opened',
    #'Replaced by', 'Website',
    #   'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
     #  'Single rider line available', 'Restraint Style',
     #  'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced', 'latitude', 'longitude', 'Type_Main',
       'opening_date_clean', 
    #'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph',
    #'height_value', 'height_unit', 
    'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()
```

```python
#pd.to_numeric(df['example:column_name_goes_here']) <<--if I want to change it to numeric Dtype
df['opening_date_clean'] = pd.to_datetime(df['opening_date_clean'])
```

```python
#renaming my columns 

df = df.rename(columns={
    'coaster_name': 'Coaster_Name',
    'year_introduced': 'Year_Introduced',
    'opening_date_clean': 'Opening_Date',
    'speed_mph': 'Speed_MPH',
    'height_ft': 'Height_FT',
    'Inversions_clean': 'Inversions',
    'Gforce_clean': 'Gforce'
})
```

```python
ax = df['Year_Introduced'].value_counts()\
.head(10)\
.plot(kind='bar', title='top 10 years roller coasters were introduced')
ax.set_xlabel('Year introduced')
ax.set_ylabel('Count')
```

```python
ax = df['Speed_MPH'].plot(kind='hist',
                     bins=20,
                     title="Roller coaster's speed (MPH)", color='pink')
ax.set_xlabel('Speed in MPH')
```

```python
ax = df['Speed_MPH'].plot(kind='kde',
                          title="Roller coaster's speed (MPH)", 
                          color='red')
ax.set_xlabel('Speed in MPH')
```

```python
df.plot(kind='scatter',
        x='Speed_MPH', 
        y='Height_FT', 
       title='Coaster speed vs Height')
plt.show()
```

```python
sns.scatterplot(x='Speed_MPH', 
        y='Height_FT',
                hue='Year_Introduced',
               data=df)
```

```python
sns.pairplot(df, vars=['Year_Introduced','Speed_MPH','Height_FT', 'Gforce'], hue='Type_Main')
plt.show()
```

```python
df_corr = df[['Speed_MPH', 'Height_FT', 'Inversions', 'Gforce']].dropna().corr()
df_corr
```

```python
sns.heatmap(df_corr, annot=True)
```

```python
df['Location'].value_counts()

ax = df.query('Location != "Other"')\
.groupby('Location')['Speed_MPH']\
.agg(['mean', 'count'])\
.query('count >= 10')\
.sort_values('mean')['mean']\
.plot(kind='barh', figsize=(12, 5), color='orange', title='Average coaster speed by location')

ax.set_xlabel('Average coaster speed')
```
